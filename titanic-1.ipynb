```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv('/kaggle/input/titanic/train.csv')
df1.info()
```

```python
df1
```

```python
df1.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis=1,inplace=True)
```

```python
df1
```

```python
df1['Age'].fillna(df1['Age'].mean(),inplace=True)
df1
```

```python
y=df1.Survived
y
```

```python
df1
```

```python
dumini=pd.get_dummies(df1.Sex)
dumini
```

```python
x=pd.concat([df1,dumini],axis=1)
x
```

```python
x.drop(['Sex','Survived'],axis=1,inplace=True)
```

```python
x.skew()
```

```python
iq1=x['Fare'].quantile(0.25)
iq2=x['Fare'].quantile(0.75)
x['Fare']=np.where(x['Fare']<iq1,iq1,x['Fare'])
x['Fare']=np.where(x['Fare']>iq2,iq2,x['Fare'])
x['Fare'].skew()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```
