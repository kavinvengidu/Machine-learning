```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/kaggle/input/spaceship-titanic/train.csv')
df
```

```python
df.info( )
```

```python
df.drop(df.iloc[:,12:13],axis=1,inplace=True)
```

```python
df
```

```python
df.drop(df.iloc[:,0:2],axis=1,inplace=True)
df
```

```python
df.drop(df.iloc[:,1:3],axis=1,inplace=True)
df
```

```python
df=df.replace(to_replace=False,value=0)
df=df.replace(to_replace=True,value=1)
df
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
df['CryoSleep'].fillna(df['CryoSleep'].mean(),inplace=True)
df['VIP'].fillna(df['VIP'].mean(),inplace=True)
df['RoomService'].fillna(df['RoomService'].mean(),inplace=True)
df['FoodCourt'].fillna(df['FoodCourt'].mean(),inplace=True)
df['ShoppingMall'].fillna(df['ShoppingMall'].mean(),inplace=True)
df['Spa'].fillna(df['Spa'].mean(),inplace=True)
df['VRDeck'].fillna(df['VRDeck'].mean(),inplace=True)
df
```

```python
x=df.iloc[:,0:8]
y=df.Transported
x
```

```python
std=x.iloc[:,3:]
std
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(std)
std1=scaler.transform(std)
std_df=pd.DataFrame(std1)
```

```python
std_df
```

```python
x=pd.concat([x,std_df],axis=1)
```

```python
x.drop(['RoomService','FoodCourt','ShoppingMall','Spa','VRDeck'],axis=1,inplace=True)
x
```

```python
y
```

```python
x.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=800)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```
